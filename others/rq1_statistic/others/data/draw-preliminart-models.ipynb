```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt
from matplotlib.pyplot import MultipleLocator
plt.rc('font',family='Times New Roman')
import matplotlib 
matplotlib.rcParams['pdf.fonttype'] = 42 
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
ft = 24
fig = plt.figure(figsize=(16,10))
ax1 = fig.add_subplot(111) 
for name,params,PFLOPs in cv_models:
    params = params * 1000000
    PFLOPs = PFLOPs * 1000000000000
    ax1.scatter(params,PFLOPs,color="darkblue",s=1200)
    ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*0.65, PFLOPs*1.7),fontsize=ft)
for name,params,PFLOPs in nlp_models:
    params = params * 1000000
    PFLOPs = PFLOPs * 1000000000000
    ax1.scatter(params,PFLOPs,color="darkred",marker="*",s=2500)
    if name=="GPT-3":
        ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*0.3, PFLOPs*0.8),fontsize=ft)
    elif name=="BERT-Base":
        ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*1.4, PFLOPs*0.8),fontsize=ft)
    elif name=="Megatron":
        ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*1.4, PFLOPs*0.8),fontsize=ft)
    elif name=="RoBERTa-Large":
        ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*0.4, PFLOPs*2.5),fontsize=ft)
    elif name=="GPT-1":
        ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*1.4, PFLOPs*0.8),fontsize=ft)
    else:    
        ax1.annotate(name, xy = (params, PFLOPs), xytext = (params*0.65, PFLOPs*2.5),fontsize=ft)
        
ax1.set_yscale('log')
ax1.set_xscale('log')

plt.ylabel("Trainning\nComplexity (Flops)",fontsize=70,labelpad=20)
plt.xlabel("# of Model Parameters",fontsize=70,labelpad=20)

plt.xticks(fontsize=60) 
plt.yticks(fontsize=60)

ax1.spines['bottom'].set_linewidth(2);###设置底部坐标轴的粗细
ax1.spines['left'].set_linewidth(2);####设置左边坐标轴的粗细
ax1.spines['right'].set_linewidth(2);###设置右边坐标轴的粗细
ax1.spines['top'].set_linewidth(2);####设置上部坐标轴的粗细

ax1.scatter(61* 1000000,460* 1000000000000,color="darkblue",s=1200,label="CV Models")
ax1.scatter(110* 1000000,6400* 1000000000000,color="darkred",marker="*",s=2500,label="NLP Models")
plt.legend(ncol=1,loc=7,fontsize=55,edgecolor="0",bbox_to_anchor=(1,0.25),labelspacing=1.5,borderpad=1.5,frameon=False)

plt.savefig('../figs/motivation-cost.pdf', bbox_inches="tight")
```
