```python
import numpy as np  
import matplotlib.pyplot as plt  
import pandas as pd
import matplotlib.dates as mdates

from matplotlib.pyplot import MultipleLocator

def plot():
    ft = 20
    lw = 4
    x1 = np.load("sorted_commit_dates.npy")
    y1 = np.load("commit_count_values.npy")
    x2 = np.load("sorted_commit_dates_code.npy")
    y2 = np.load("commit_count_values_code.npy")

    fig, ax1 = plt.subplots(figsize=(10,5), dpi=100)
    ax1.plot(np.array(x1), y1, linewidth = lw, color='darkblue', linestyle = "--", label="Number of Commits")

    ax=plt.gca()
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))  
    ax.xaxis.set_major_locator(mdates.MonthLocator(interval=6))  
    ax.yaxis.set_major_locator(MultipleLocator(400))

    plt.tick_params(axis='y',colors='darkblue')
    plt.xticks(fontsize=ft, rotation=45)
    plt.yticks(fontsize=ft)
    ax1.set_ylabel("Number of Commits",  color='darkblue', fontsize=ft)
    ax1.set_xlabel('Time',  fontsize=ft)
    plt.legend(fontsize=ft, loc=3, bbox_to_anchor=(0.05, 1.03), ncol=1,borderaxespad = 0., frameon=False,handletextpad=0.3)

    ax2 = ax1.twinx()  # this is the important function
    ax2.plot(np.array(x2), y2, 'r', linewidth = lw, color="darkred", label="Lines of Merged Code")

    ax=plt.gca()
    ax.yaxis.set_major_locator(MultipleLocator(20000))

    plt.tick_params(axis='y',colors='darkred')
    plt.xticks(fontsize=ft)
    plt.yticks(fontsize=ft)
    ax2.set_ylabel("Lines of Merged Code", color='darkred',fontsize=ft)


    plt.legend(fontsize=ft, loc=3, bbox_to_anchor=(0.5, 1.03), ncol=1,borderaxespad = 0., frameon=False,handletextpad=0.3)
    plt.savefig("./figure6.pdf", bbox_inches="tight")

plot()
```
